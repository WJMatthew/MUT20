## Viewing our data

```python
import os
import pandas as pd


directory = 'data'

dfs = {}

for file in os.listdir(directory):
    if '.csv' in file:
        print(directory, file)
        dfs[file.rstrip('.csv')] = pd.read_csv(os.path.join(directory, file))
```

```python
def filter_df(df, pos, sort_by=None):
    result = df[ df['position'] == pos]
    if sort_by:
        return result.sort_values(sort_by, ascending=False)
    return result

#dfs['QB'].sort_values('OVR', ascending=False).head(10)

filter_df(dfs['OFF'], 'HB', 'BTK')[['name', 'BTK', 'SPD', 'TRK', 'AGI']]
```
